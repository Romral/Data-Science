<h1>MINOR PROJECT</h1>

<h3>Objective of this Analysis:</h3>

To understand how the student's performance (test scores) is affected by the other variables (Gender, Ethnicity, Parental level of education, Lunch, Test preparation course).

gender: Gender of the student

race/ethnicity: Race of the Student As Group A/B/C

parental level of education: What is the education Qualification of Student Parent

lunch:Whether the lunch is Standard type/Free lunch or Some Discounted Lunch.

test preparation course : Whether Student has Taken or not and Completed.

math score : Scores in Maths

reading score : Scores in Reading

writing score : Scores in Writing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<h2>CountPlot for Categorical variables:</h2>

```python
cat_var = ['gender','race/ethnicity','parental level of education','lunch','test preparation course']
for i in cat_var:
    plt.figure(figsize=(10,6))
    sns.set_style("darkgrid")
    sns.countplot(x=i,data=file)
    plt.title("Countplot of "+i,fontsize=15)
    plt.show()
```

<h3>Observations:</h3>

In Gender we analyzed that the count of female is more than male.

In Race/Ethnicity we analyzed that the count of group A is least whereas group C is highest.

In Parental Level Of Education we analyzed that the count of Master's Degree is least whereas some college is highest.

In Lunch we analyzed that the count of Standard lunch is more than Free/reduced.

In Test preparation course we analyzed that the count of none is more than completed.

<h2>Variation of scores with race/ethnicity in terms of genders:</h2>

```python
scores = ['math score','reading score','writing score']
for i in scores:
    plt.figure(figsize=(10,6))
    sns.barplot(x = 'race/ethnicity',y = i,hue = 'gender',data = file)
    plt.title("Variation of "+i+" with race/ethnicity",fontsize = 15)
    plt.show()
```

<h3>Observations:</h3>

Group E seems to have the highest scores for all the scores especially the math score.For reading and writing the scores seem to not differ by a lot. 

Males scored more in math for all groups of Race/Ethnicity.

Females scored more in reading and writing for all groups of Race/Ethnicity.

<h3>Inference:</h3>

Gender has chances of being an important variable for determining the scores.

Race/Ethnicity has chances of holding some importance when it comes to scores.

<h2>Variation of scores with parent level of education in terms of lunch:</h2>

```python
scores = ['math score','reading score','writing score']
for i in scores :
    plt.figure(figsize=(10,6))
    sns.barplot(y = i,x='parental level of education',hue='lunch',data=file)
    plt.title("Variation of "+i+" with parental level of education",fontsize = 15)
    plt.show()
```

<h3>Observations:</h3>

The standard lunch is dominating in all types of parental level of education for all the 3 scores.

There is no extreme or clear distinction in scores for each category of Parental Level of Education as the scores don't differ a lot category-wise 

<h3>Inference:</h3>

Lunch has high chances of being an important variable for determining the scores.

Parental Level of Education has chances of not being an Important Variable.
